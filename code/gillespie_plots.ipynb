```python
%matplotlib inline
import csv
from __future__ import division
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt


t,ign,spr,hib,sti = [], [], [], [], []
with open('crit_thresh', 'r') as f:
    reader = csv.reader(f,delimiter=',')
    f.readline()
    for row in reader:
        t.append(float(row[0]))
        ign.append(float(row[1]))
        spr.append(float(row[2]))
        #hib.append(float(row[3]))
        sti.append(float(row[3])+float(row[4]))
        
        
# plot 
f = plt.figure();    
plt.plot(t,ign, 'b.', label=' ignorants');
plt.plot(t,spr, 'r.',  label='spreaders');
#plt.plot(t,hib, 'c.', label='hibernators');
plt.plot(t,sti, 'g.', label='stiflers');
plt.xlim(0,10000000)
plt.ylim(0,1)
plt.legend(loc='best')
plt.xlabel("time", fontsize=20);
plt.ylabel("Fraction of population", fontsize=20);
plt.savefig("crit_theshold.png")
plt.show()
```
